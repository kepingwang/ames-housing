```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../data/data_ready.csv')
df.head()
dfX = df.drop('SalePrice', axis=1)
y = df.SalePrice.values
X = df.drop('SalePrice', axis=1).values
print("X shape: {}".format(X.shape))
print("y shape: {}".format(y.shape))
```

```python
dfX
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
pca
# print(pca.components_) 
```

```python
n_components = 2

pca = PCA(n_components=5)
pca.fit(X)
print(pca.explained_variance_ratio_)
for pcID in range(n_components):
  print("principle component {}, explained variance {}".format(pcID+1, pca.explained_variance_ratio_[pcID]))
  for field_index in pca.components_[pcID].argsort()[::-1][:6]:
    print(dfX.columns[field_index], pca.components_[pcID][field_index])
```

```python
X_reduced = pca.transform(X)
X_reduced.shape
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
def rmse(y, y_pred):
  return np.sqrt(((y - y_pred)**2).mean())
def evaluate(y, y_pred):
  print("RMSE: {}".format(rmse(y, y_pred)))
  sse = ((y - y_pred)**2).sum()
  sst = ((y - y.mean())**2).sum()
  r2 = 1.0 - sse / sst
  print("R2: {}".format(r2))
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf', C=1e4, epsilon=0.003) # large C means small margin
svr.fit(X, y)
y_pred = svr.predict(X)
```

```python
evaluate(y, y_pred)
```

```python
from sklearn.neural_network import MLPRegressor
nn = MLPRegressor(hidden_layer_sizes=(3,), max_iter=200, alpha=0.0001,learning_rate_init=0.001)
nn.fit(X, y)
y_pred = nn.predict(X)
```

```python
evaluate(y, y_pred)
```
